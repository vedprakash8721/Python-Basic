2. What is an interpreter in Python ?
-   An interpreter in Python is a program that reads and executes Python code line by line. Instead of converting the whole program into machine code at once like a compiler, the interpreter runs each line one at a time, which makes it easier to test and debug your code. So when you write and run a Python script, the interpreter is the one actually reading your code and making it work immediately.

3. What are pre-defined keywords in Python ?
-   Pre-defined keywords in Python are special words that are reserved by the language because they have specific meanings and functions. You can’t use them as variable names, function names, or identifiers. These keywords are part of the syntax and structure of Python—for example, if, else, while, for, def, return, and class. They help Python understand what kind of action you want your code to perform.

4. Can keywords be used as variable names ?
-   No, keywords cannot be used as variable names in Python. Since keywords have special meanings in the language, using them as variable names would confuse the interpreter. For example, trying to name a variable if or class will result in a syntax error because Python expects those words to be used for control flow or defining classes, not for storing values.

5. What is mutability in Python ?
-   Mutability in Python refers to whether or not an object can be changed after it is created. If an object is mutable, you can change its content without changing its identity. Examples of mutable objects include lists, dictionaries, and sets. If an object is immutable, it cannot be changed once it's created—any modification results in a new object. Examples of immutable objects are strings, integers, and tuples. Understanding mutability helps you avoid bugs when passing data around in your programs.
6.  Why are lists mutable, but tuples are immutable ?
-    Lists are mutable because they are designed to store a collection of items that might change over time—you can add, remove, or modify elements in a list after it's created. Tuples, on the other hand, are immutable by design to ensure that their contents remain constant. This immutability makes tuples more memory-efficient and safer to use when you need fixed data, like as keys in dictionaries or for returning multiple values from a function without risking accidental changes.

-    In Python, == checks whether the values of two variables are the same, while is checks whether they refer to the exact same object in memory.

9. What is type casting in Python ?
-   Type casting in Python refers to converting one data type into another. Python allows you to change a variable's type using explicit type conversion (also called type casting) or implicit type conversion (done automatically by Python).

Explicit Type Casting: This is when you manually convert one type to another using functions like:

int() to convert to an integer

float() to convert to a floating-point number

str() to convert to a string

list(), tuple(), set(), etc., for converting between collection types.
x = "123"  # string

- Implicit Type Casting: This happens automatically when Python converts a smaller data type to a larger one, like converting an integer to a float.

Example:
x = 5  # integer
y = 2.5  # float
result = x + y  # Python automatically converts x to float
print(result)  # 7.5

12. How does the elif statement work ?
-   The elif (short for "else if") statement in Python allows you to check multiple conditions in a sequence. It provides an additional condition to check after the initial if condition. If the if condition is false, Python evaluates the elif conditions one by one. If one of the elif conditions is true, its corresponding block of code is executed. If none of the if or elif conditions are true, the else block (if present) is executed.

Here's how it works:

The program first checks the condition in the if statement.

If the if condition is True, its block of code is executed, and the remaining elif or else statements are skipped.

If the if condition is False, the program checks the conditions in the elif statements, one by one.

If any elif condition is True, the corresponding block of code is executed, and the rest of the conditions are skipped.

If none of the conditions are True, the code in the else block is executed (if present).

x = 10

if x > 15:
    print("x is greater than 15")
elif x == 10:
    print("x is equal to 10")
elif x < 5:
    print("x is less than 5")
else:
    print("x is less than 15 but greater than 5")


13. What is the difference between for and while loops ?
-   for loop:

It is typically used when the number of iterations is known or predetermined.

The loop iterates over a sequence (like a list, tuple, string, or range) or any iterable object.

The syntax is more concise and easy to use when dealing with a sequence of items.

Example:
for i in range(5):  # Loops from 0 to 4
    print(i)

while loop:

It is used when the number of iterations is not known in advance and depends on a condition being true or false.

The loop continues to execute as long as the specified condition is True. If the condition is initially False, the loop doesn’t execute at all.

Example:
i = 0
while i < 5:  # Continues as long as i is less than 5
    print(i)
    i += 1

14. Describe a scenario where a while loop is more suitable than a for loop.?
A while loop is more suitable than a for loop in scenarios where the number of iterations is not known in advance and is based on a condition that needs to be evaluated dynamically during execution.
Scenario Example: User Login Attempt Imagine a scenario where a user is trying to log into an account, and you want to allow a maximum of 3 attempts to enter the correct password. If the user enters the wrong password, the program should continue to prompt for the password until the correct one is entered or the number of attempts exceeds 3.

Since the number of attempts is not predetermined, a while loop is more suitable because it can continue running as long as the condition (number of attempts < 3) is true.

Example Code:
correct_password = "password123"
attempts = 0

while attempts < 3:
    user_input = input("Enter your password: ")
    if user_input == correct_password:
        print("Login successful!")
        break
    else:
        print("Incorrect password.")
        attempts += 1

if attempts == 3:
    print("Too many failed attempts. Please try again later.")
